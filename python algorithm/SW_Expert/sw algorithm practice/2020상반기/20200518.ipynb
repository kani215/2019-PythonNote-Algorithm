```python
from collections import deque 
import copy
# 2-1. DFS ----------------------------------
def DFS(start) :
    graph2 = copy.deepcopy(graph)
    switchclock = [0,0,0,0]
    dirction =[0,0,0,0]
    deque3 = deque()
    deque3.append(start)
    x1,y1,x2,y2, dirt1 = deque3.pop()
    deque3.append((x1,y1,dirt1))
    deque3.append((x2,y2,dirt1))
    while deque3 :
        x,y,dirt = deque3.pop()
        for i in range(2) :
            nx = x+dx[i]
            ny = y+dy[i]
            if nx >= 0 and nx<4 and ny >=0 and ny<8 and graph2[nx][ny] != graph2[x][y] :
                graph2[nx][ny] = graph2[x][y]
                dirt2 = 0
                if dirt == 1 :
                    dirt2 = -1
                else :
                    dirt2 = 1
                switchclock[x] = 1
                switchclock[nx] = 1
                dirction[x] = dirt
                dirction[nx] = dirt2
                deque3.append((nx,2,dirt2))
                deque3.append((nx,6,dirt2))
        
    for i in range(4) :
        if switchclock[i] :
            move(i,dirction[i])
            
# 2-2. 톱니바퀴 이동 함수
def move(nclock,dirct) :
    global graph
    if dirct == 1 :
        a = graph[nclock].pop(-1)
        graph[nclock] = [a] + graph[nclock]
    else :
        a = graph[nclock].pop(0)
        graph[nclock].append(a)


# 1. for문 연계 -----------------------------
graph = [0] * 4
dx = [-1,1]
dy = [-4,4]

for i in range(4) :
    temp =list(input())
    graph[i] = [int(x) for x in temp]
for j in range(int(input())) :
    nclock,dirct = map(int,input().split())
    DFS((nclock-1,2,nclock-1,6,dirct))
    
    # 방향 수는 0 -1 1 3개가 존재
    
# 3. 점수 환산 --------------------------------
sum = 0 
for q in range(4) :
    sum += (2**q) * graph[q][0]
print(sum)
```
