```python
def merge_sort(a) :
    n =len(a)
    if n<=1 :
        return a
    # 재귀상태 종료 조건이자 단일 원소 리스트 거부 조건일까... 1이면 그 리스트를 반환해줍니다.
    mid =n//2
    g1 = merge_sort(a[:mid])
    g2 = merge_sort(a[mid:])
    print("g1:",g1,"g2:",g2,"a:",a,"n:",n)
    # 재귀 호출로 정렬 

    result = []
    while g1 and g2 :  # while +리스트 direct는 리스트가 빈값이 될때까지 반복이라는 의미
        if g1[0] < g2[0]:
            print("resultif:",result,"g1:",g1,"g2:",g2,"a:",a)
            result.append(g1.pop(0))
            print("resultif:",result,"g1:",g1,"g2:",g2,"a:",a)
        else : 
            print("resultes:",result,"g1:",g1,"g2:",g2,"a:",a)
            result.append(g2.pop(0))
            print("resultes:",result,"g1:",g1,"g2:",g2,"a:",a)

    # 둘중 하나가 비어버리면 위의 while문이 중단됨 그 경우에는 밑의 while문이 돌아간다.
    # 나머지 떨이를 그대로 추가만 진행해줌
    while g1 :
        result.append(g1.pop(0))
        print("resultwh:",result,"g1:",g1,"g2:",g2,"a:",a)

    while g2:
        result.append(g2.pop(0))
        print("resultwh:",result,"g1:",g1,"g2:",g2,"a:",a)
        
    return result
d1 = [7]
d2 = [1,32,3]
d = [6,8,3,9,10,1,2,4,7,5]
print(merge_sort(d2))
```

# 카운팅 정렬 

```python
def counting_sort(A):
    if min(A) < 0:
        raise
    i, n, k = 0, len(A), max(A)+1
    C = [0]*k
    for j in range(n):
        C[A[j]] = C[A[j]]+1
    for j in range(k):
        while C[j] > 0:
            (A[i], C[j], i) = (j, C[j]-1,i+1)
```

N개의 수가 주어졌을 때, 이를 오름차순으로 정렬하는 프로그램을 작성하시오.

입력
첫째 줄에 수의 개수 N(1 ≤ N ≤ 1,000,000)이 주어진다. 둘째 줄부터 N개의 줄에는 숫자가 주어진다. 이 수는 절댓값이 1,000,000보다 작거나 같은 정수이다. 수는 중복되지 않는다.

출력
첫째 줄부터 N개의 줄에 오름차순으로 정렬한 결과를 한 줄에 하나씩 출력한


```python
def merge_sort(a) :
    n =len(a)
    if n<=1 :
        return a

    mid = n//2
    g1 = merge_sort(a[:mid])
    g2 = merge_sort(a[mid:])
    result = [] 
    
    try :
        for i in range(n): 
            if g1[0] < g2[0]:
                result.append(g1[0])
                del g1[0]

            else : 
                result.append(g2[0])
                del g2[0]

    except : 
            if len(g1)==0:
                for i in range(len(g2)) :
                    result.append(g2[i])
            else :
                for i in range(len(g2)) :
                    result.append(g1[i])
    return result

import sys
T = int(input())
emplist =[]
for i in range(T) :
    test_case = int(input())
    emplist.append(test_case)
    
k = merge_sort(emplist)
    
for j in range(len(k)) :
    print(k[j])
```

```python
def counting_sorted(arr, K):
    c = [0] * T
    sorted_arr = [0] * len(arr)
    
    for i in arr:
        c[i] += 1
    
    for i in range(1,K):
        c[i] += c[i-1]
   
    for i in range(len(arr)):
        sorted_arr[c[arr[i]]-1] = arr[i]
        c[arr[i]] -= 1

    return sorted_arr
T = int(input())
arr = [int(input()) for _ in range(T)]
print(counting_sorted(arr, T))
```
